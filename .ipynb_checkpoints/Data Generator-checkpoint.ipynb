```python
import csv
import numpy as np
import math
import random

def class_compare(a):
    if (int(a[1]) < 3) or (int(a[2]) > 1024) and (int(a[3]) > 14):
          return True        
    return False

data_count = 1000

data_list = [['priority', 'c', 0, 10, 5], ['data length', 'n', 512], ['power of nodes', 'n', 10]]

class_list = ['5g', 'lora']

classifier_probability = 0.6

#prepare output file
file = open('output.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(file)

#normal distribution parameters - for numerical data
mu, sigma = 0, 1

label = ['index']

i = 0
while i < len(data_list):
    if (data_list[i][1] == 'n'):
        #numerical data - use normal distribution for naive data generation
        dist = np.random.normal(data_list[i][2], data_list[i][2] * sigma, data_count)
        #calcaulate values for nonzero
        j = 0
        while j < len(dist):
            prob = 1
            if dist[j] < 0:
                dist[j] *= -1
            j += 1
        
        
    else:
        #categorical data - use uniform distribution for naive data generation
        dist = np.random.uniform(low=data_list[i][2], high=data_list[i][3], size = data_count).round(0)
        
    data_list[i].append(dist)
    label.append(data_list[i][0])
    i += 1

label.append('class')
wr.writerow(label)

i = 0
while i < data_count:
    j = 0 
    line = [i]
    while j < len(data_list):
        value = data_list[j][len(data_list[j])-1][i]
        line.append(value)
        j+=1
        
    if class_compare(line):
        line.append(class_list[0])
    else:
        line.append(class_list[1])
    wr.writerow(line)       
    i+=1

file.close()
```
